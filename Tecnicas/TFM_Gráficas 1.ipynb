```python
import os
import pandas as pd
data = pd.read_csv("TFM_limpio.csv",delimiter=",")
data
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convertir la columna 'Start_Time' a datetime especificando el formato
data['Start_Time'] = pd.to_datetime(data['Start_Time'], format='%Y-%m-%d %H:%M:%S')

# Luego puedes proceder con tu cÃ³digo para extraer el aÃ±o y el mes
data['Year_Month'] = data['Start_Time'].dt.to_period('M')
accidents_per_month = data['Year_Month'].value_counts().sort_index()

# Graficar la cantidad de accidentes por mes
plt.figure(figsize=(8, 4))
accidents_per_month.plot()
plt.title('Cantidad de accidentes por mes')
plt.xlabel('Mes')
plt.ylabel('Cantidad de accidentes')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponiendo que tienes un DataFrame llamado 'data' que contiene tus datos

# Extraer el aÃ±o de la fecha de inicio
data['Year'] = data['Start_Time'].dt.year

# Contar la cantidad de accidentes por aÃ±o
accidents_per_year = data['Year'].value_counts()

# Ordenar los valores por aÃ±o (de menos a mÃ¡s)
accidents_per_year_sorted = accidents_per_year.sort_index()

# Graficar la cantidad de accidentes por aÃ±o
plt.figure(figsize=(8, 4))
accidents_per_year_sorted.plot(kind='bar')
plt.title('Cantidad de accidentes por aÃ±o')
plt.xlabel('AÃ±o')
plt.ylabel('Cantidad de accidentes')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convertir la columna 'Start_Time' a datetime utilizando el formato 'ISO8601' o 'mixed'
data['Start_Time'] = pd.to_datetime(data['Start_Time'], format='ISO8601', errors='coerce')

# Extraer el aÃ±o y el mes de la fecha de inicio
data['Year'] = data['Start_Time'].dt.year
data['Month'] = data['Start_Time'].dt.month

# Filtrar los datos para incluir solo los meses de enero
data_jan = data[(data['Month'] == 1) & (data['Year'] != 2016)]

# Contar la cantidad de accidentes por aÃ±o para el mes de enero
accidents_per_year_jan = data_jan['Year'].value_counts()

# Ordenar los valores por aÃ±o (de menos a mÃ¡s)
accidents_per_year_jan_sorted = accidents_per_year_jan.sort_index()

# Graficar la cantidad de accidentes por aÃ±o para el mes de enero
plt.figure(figsize=(8, 4))
accidents_per_year_jan_sorted.plot(kind='bar', color='skyblue')
plt.title('Cantidad de accidentes en enero por aÃ±o (excluyendo 2016)')
plt.xlabel('AÃ±o')
plt.ylabel('Cantidad de accidentes')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Ordenar los datos por fecha
data.sort_values(by='Start_Time', inplace=True)

# Contar la cantidad de accidentes por dÃ­a
accidents_per_day = data['Start_Time'].dt.date.value_counts().sort_index()

# Calcular el aumento porcentual consecutivo de los accidentes
relative_increase = accidents_per_day.pct_change() * 100
cumulative_increase = relative_increase.cumsum()

# Graficar el aumento porcentual consecutivo de los accidentes
plt.figure(figsize=(8, 4))
plt.plot(relative_increase, color='skyblue', label='Aumento relativo (%)')
plt.plot(cumulative_increase, color='red', label='Aumento acumulado (%)')
plt.title('Aumento porcentual consecutivo de accidentes')
plt.xlabel('Fecha')
plt.ylabel('Porcentaje de aumento')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Extraer la hora del dÃ­a de la columna 'Start_Time'
data['Hour'] = data['Start_Time'].dt.hour

# Contar la cantidad de accidentes por hora del dÃ­a
accidents_per_hour = data['Hour'].value_counts().sort_index()

# Graficar la cantidad de accidentes por hora del dÃ­a
plt.figure(figsize=(8, 4))
accidents_per_hour.plot(kind='bar', color='skyblue')
plt.title('Cantidad de accidentes por hora del dÃ­a')
plt.xlabel('Hora del dÃ­a')
plt.ylabel('Cantidad de accidentes')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Suponiendo que tienes un DataFrame llamado 'data' que contiene tus datos

# Convertir la columna 'Start_Time' a datetime si no lo estÃ¡ ya
data['Start_Time'] = pd.to_datetime(data['Start_Time'], format='ISO8601', errors='coerce')

# Extraer la hora del dÃ­a de la columna 'Start_Time'
data['Hour'] = data['Start_Time'].dt.hour

# Contar la cantidad de accidentes por hora del dÃ­a segÃºn la severidad
accidents_per_hour_severity = data.groupby(['Hour', 'Severity']).size().unstack(fill_value=0)

# Graficar la cantidad de accidentes por hora del dÃ­a segÃºn la severidad
accidents_per_hour_severity.plot(kind='bar', stacked=True, figsize=(8, 4), color=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Cantidad de accidentes por hora del dÃ­a segÃºn severidad')
plt.xlabel('Hora del dÃ­a')
plt.ylabel('Cantidad de accidentes')
plt.xticks(rotation=0)
plt.legend(title='Severidad')
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convertir la columna 'Start_Time' a datetime si aÃºn no lo es
data['Start_Time'] = pd.to_datetime(data['Start_Time'])

# Obtener el dÃ­a de la semana y contar la cantidad de accidentes por dÃ­a
accidents_per_weekday = data['Start_Time'].dt.dayofweek.value_counts().sort_index()

# Mapear los nÃºmeros de los dÃ­as de la semana a los nombres correspondientes
weekday_names = ['Lunes', 'Martes', 'MiÃ©rcoles', 'Jueves', 'Viernes', 'SÃ¡bado', 'Domingo']
accidents_per_weekday.index = [weekday_names[idx] for idx in accidents_per_weekday.index]

# Graficar la cantidad de accidentes por dÃ­a de la semana
plt.figure(figsize=(8, 4))
accidents_per_weekday.plot(kind='bar')
plt.title('Cantidad de accidentes por dÃ­a de la semana')
plt.xlabel('DÃ­a de la semana')
plt.ylabel('Cantidad de accidentes')
plt.show()
```

```python
# Iterar sobre las columnas categÃ³ricas y crear una grÃ¡fica para cada una
# Definir las columnas categÃ³ricas
categorical_columns = [ 'Weather_Condition', 'Timezone', 'City', 'State', 'County']

for column in categorical_columns:
    # Obtener el conteo de valores
    value_counts = data[column].value_counts()
    
    # Seleccionar los 10 valores mÃ¡s frecuentes
    top_10_values = value_counts.head(10)
    
    # Configurar la figura
    plt.figure(figsize=(7, 5))
    
    # Graficar la proporciÃ³n de los 10 valores mÃ¡s frecuentes
    top_10_values.plot(kind='bar', color='skyblue')
    
    # Configurar tÃ­tulo y etiquetas
    plt.title(f"Top 10 valores mÃ¡s frecuentes en '{column}'")
    plt.xlabel('Valor')
    plt.ylabel('Frecuencia')
    plt.xticks(rotation=45)  # Rotar las etiquetas del eje x para mayor legibilidad
    
    # Mostrar la grÃ¡fica
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Excluir la variable ID y otras no numÃ©ricas
data_numeric = data.select_dtypes(include=['number'])

# Calcular estadÃ­sticas descriptivas de las variables numÃ©ricas
numeric_stats = data_numeric.describe()

# Imprimir estadÃ­sticas descriptivas
print("EstadÃ­sticas descriptivas de variables numÃ©ricas:")
print(numeric_stats)

# Calcular la correlaciÃ³n entre las variables numÃ©ricas
correlation_matrix = data_numeric.corr()

# Eliminar la variable 'Severity' de la matriz de correlaciÃ³n
if 'Severity' in correlation_matrix.columns:
    correlation_matrix = correlation_matrix.drop('Severity', axis=1)
    correlation_matrix = correlation_matrix.drop('Severity', axis=0)

# Graficar la matriz de correlaciÃ³n
plt.figure(figsize=(5, 3))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title('Matriz de correlaciÃ³n entre variables numÃ©ricas')
plt.show()
```
